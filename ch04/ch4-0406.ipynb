```python
error = 0
str1 = input(":")
print("",str1)
while error<5:
    strA = "-"+str1[-1]+"-:"
    str2=input(strA)
    if str1[-1] == str2[0]:
        str1 = str1 + "-" + str2 
        print("",str1)
    else :
        error += 1
print("")
```
